```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

```

### Reading df

```python

df = pd.read_csv('ASE_info.csv')
df.index = df['Gene']
df
```

### Cleaning DF

```python
colums_to_drop = [col for col in df.columns if not re.search(r"TPM", col) or re.search(r"total", col)]
cleaned_df = df.drop(colums_to_drop, axis=1)
cleaned_df = cleaned_df.dropna(how='all', axis=0)
cleaned_df.shape
```
